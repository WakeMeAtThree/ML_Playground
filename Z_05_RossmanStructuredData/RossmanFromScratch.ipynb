# Extract Features

```python
for df in (joined,joined_test):
    param = dict(year=df['CompetitionOpenSinceYear'],month=df['CompetitionOpenSinceMonth'],day=15)
    df['CompetitionOpenSince'] = pd.to_datetime(param)
```

```python
for df in (joined,joined_test):
    df['CompetitionDaysOpen'] = df.Date.subtract(df['CompetitionOpenSince']).dt.days
```

```python
for df in (joined, joined_test):
    df.loc[df.CompetitionDaysOpen<0,'CompetitionDaysOpen'] = 0
    df.loc[df.CompetitionOpenSinceYear<1990,'CompetitionDaysOpen'] = 0
```

```python
def constrain(val, R1, R2):
    if(val > R2): return R2
    if(val < R1): return R1
    return val
```

```python
for df in (joined, joined_test):
    df['CompetitionMonthsOpen']=(df['CompetitionDaysOpen']//30).apply(lambda x: constrain(x,0,24))
```

```python
joined.Promo2SinceWeek = joined.Promo2SinceWeek.astype(int)
joined.Promo2SinceYear = joined.Promo2SinceYear.astype(int)
```

```python
joined_test.Promo2SinceWeek = joined_test.Promo2SinceWeek.astype(int)
joined_test.Promo2SinceYear = joined_test.Promo2SinceYear.astype(int)
```

```python
for df in (joined,joined_test):
    df["Promo2Since"] = pd.to_datetime(df.apply(lambda x: Week(
        x.Promo2SinceYear, x.Promo2SinceWeek).monday(), axis=1).astype(pd.datetime))
    df["Promo2Days"] = df.Date.subtract(df["Promo2Since"]).dt.days
```

```python
for df in (joined,joined_test):
    df.loc[df['Promo2Days']<0,"Promo2Days"] = 0
    df['Promo2Weeks']=(df['Promo2Days']//7).apply(partial(constrain,R1=0,R2=25))
```

```python
def get_elapsed(fld, pre, df):
    day1 = np.timedelta64(1, 'D')
    last_date = np.datetime64()
    last_store = 0
    res = []

    for s,v,d in zip(df.Store.values,df[fld].values, df.Date.values):
        if s != last_store:
            last_date = np.datetime64()
            last_store = s
        if v: last_date = d
        res.append(((d-last_date).astype('timedelta64[D]') / day1))
    df[pre+fld] = res
```
